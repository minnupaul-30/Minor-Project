```python

for filename in os.listdir('test/'):
    filename_1= filename.split('_')
    final = filename_1[1]

    print(final)
    os.rename(os.path.join('test/',filename),os.path.join('test/',final))
```

```python
for filename in os.listdir('sina//'):
    name_len = len(os.path.splitext(filename)[0])
    try:
        os.mkdir("sina/{}".format(str(name_len)))
    except FileExistsError:
        pass
    os.rename("sina/{}".format(filename), "sina/{}/{}".format(name_len,filename))
```
